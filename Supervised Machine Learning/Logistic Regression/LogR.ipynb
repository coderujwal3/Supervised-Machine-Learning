```python
# ------------------------- Example of Hours of Study and Result (Pass/Fail) -------------------------
X = [[1],[2],[3],[4],[5]]
y = [0,0,1,1,1]     # 0: fail & 1: pass

lr = LogisticRegression()
lr.fit(X, y)


# Metrics part - Linear Regression
predicted_score1 = lr.predict(X)
mae = mean_absolute_error(y, predicted_score1)
mse = mean_squared_error(y, predicted_score1)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicted_score1)
print(f"Mean Absolute Error (MAE - LogisticRegression): {round(mae,2)}")
print(f"Mean Squared Error (MSE - LogisticRegression): {round(mse, 2)}")
print(f"Root Mean Squared Error (RMSE - LogisticRegression): {round(rmse, 2)}")
print(f"R2 Score (R2 - LogisticRegression): {round(r2, 2)}")

# scatter plot & regression line
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.scatter(X, y, color='lightgreen', label='Actual Scores')
plt.plot(X, predicted_score1, color="red", label="predicted_score (Regression Line)")
plt.title("Model Prediction v/s Actual Score - Linear Regression")
plt.xlabel("hours")
plt.ylabel("fail-pass")
plt.grid(True)
plt.show()


inp = float(input("Enter the number of hours you studied: "))
out = lr.predict([[inp]])[0]

if out == 1:
    print(f'If your study {inp} hours you might get Pass')
else:
    print(f'If your study {inp} hours you might get Fail')
```
