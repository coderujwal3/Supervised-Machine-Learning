```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('Student_Marks.csv', index_col=False)
print(df)
```

```python
X = df[['time_study']].values
y = df['Marks']

model = LinearRegression()
model.fit(X,y)
predicted_score = model.predict(X)

# metrics
mae = mean_absolute_error(y, predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicted_score)

print(f"Mean Absolute Error: {round(mae,2)}")
print(f"Mean Squared Error: {round(mse,2)}")
print(f"Root Mean Squared Error: {round(rmse,2)}")
print(f"Model Accuracy (R^2): {round(r2,3)}")


# histogram data visualisation
plt.figure(figsize=(10,6))
plt.hist(df['Marks'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Exam Score")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of students")
plt.grid(True)
plt.show()



# Scatter plot + Regression Line
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='lightgreen', label='Actual Scores')
plt.plot(X, predicted_score, color="red", label="predicted_score (Regression Line)")
plt.title("Model Prediction v/s Actual Score")
plt.xlabel("Study Hours")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()



# another input value
new_hours = 9
predicted_new_score = model.predict([[new_hours]])[0]
print(f"Prediction of makrs for {new_hours} hours is: {round(predicted_new_score, 4)} marks")
```
